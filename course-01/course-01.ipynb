```python
#inversion counter

A = [1,4,5,3,2,6]

def mergesort(array):
    
    if(len(array)==1):
        
        fin = []
        fin.append(array)
        fin.append(0)
        return fin
    
    x = mergesort(array[:len(array)//2])
    y = mergesort(array[len(array)//2:])
    
    a = x[0]
    b = y[0]
    count = x[1] + y[1]
    
    final = []
    i=j=0
    
    while(len(final)<len(array)):
    
        if((i<len(a))and(j<len(b))):
        
            if(a[i]<=b[j]):

                final.append(a[i])
                i+=1

            else:

                final.append(b[j])
                j+=1
                count+=len(a)-i
            
        elif(i<len(a)):
            
            final = final + a[i:]
            
        else:
            
            final = final + b[j:]

    fin = []
    fin.append(final)
    fin.append(count)
    
    return fin

B = mergesort(A)
print(B)
```

```python
#quick sort

def quickfirst(array,low,high):
    
    if(len(array[low:high])<=1):
        return 0
    
    temp=0
    piv=array[low]
        
    i=j=low+1
    
    c=high-low-1
    
    while(j<high):
        
        if (array[j]>piv):
            j+=1
            
        else:
            temp=array[j]
            array[j]=array[i]
            array[i]=temp
            
            j+=1
            i+=1
            
    temp=array[low]
    array[low]=array[i-1]
    array[i-1]=temp
    
    a=quickfirst(array,low,i-1)
    b=quickfirst(array,i,high)
    
    return a+b+c
```

```python
def quickmed(array,low,high):
    
    if(len(array[low:high])<=1):
        return 0
    
    temp=0
    med=[]
    med.append(array[low])
    med.append(array[high-1])
    if((high+low)%2!=0):
        medsel=(high+low)//2
    else:
        medsel=(high+low-1)//2
    med.append(array[medsel])
    med.sort()
    piv=med[1]
    
    if(piv==array[low]):
        
        piv=array[low]
        
    elif(piv==array[high-1]):
    
        temp=array[high-1]
        array[high-1]=array[low]
        array[low]=temp
        
    else:
        
        temp=array[medsel]
        array[medsel]=array[low]
        array[low]=temp
        
    i=j=low+1
    
    c=high-low-1
    
    while(j<high):
        
        if (array[j]>piv):
            j+=1
            
        else:
            temp=array[j]
            array[j]=array[i]
            array[i]=temp
            
            j+=1
            i+=1
            
    temp=array[low]
    array[low]=array[i-1]
    array[i-1]=temp
    
    a=quickfirst(array,low,i-1)
    b=quickfirst(array,i,high)
    
    return a+b+c
```

```python
def quicklast(array,low,high):
    
    if(len(array[low:high])<=1):
        return 0
    
    temp=0
    piv=array[high-1]
    
    temp=array[high-1]
    array[high-1]=array[low]
    array[low]=temp
        
    i=j=low+1
    
    c=high-low-1
    
    while(j<high):
        
        if (array[j]>piv):
            j+=1
            
        else:
            temp=array[j]
            array[j]=array[i]
            array[i]=temp
            
            j+=1
            i+=1
            
    temp=array[low]
    array[low]=array[i-1]
    array[i-1]=temp
    
    a=quickfirst(array,low,i-1)
    b=quickfirst(array,i,high)
    
    return a+b+c
```

```python
#karger min cut contraction

def karger(graph):
    import random as rd
    mincut=len(graph)
    b=0

    while(b<100):
        gr=graph
        ed=list(range(0,len(graph)))
        a=0

        while (a<len(gr)-2):

            s=d=0

            s=rd.choice(ed)
            d=rd.choice(gr[s])

            gr[s].remove(d)

            gr[s].extend(gr[d])

            gr[d]=[]
            ed.remove(d)

            for i in gr:
                for j in range(0,len(i)):
                    if(i[j]==d):
                        i[j]=s

            while s in gr[s]:
                gr[s].remove(s)

            a+=1

        for i in range(0,len(gr)):
            if(gr[i]):
                
                if(mincut>len(gr[i])):
                    mincut=len(gr[i])
                    break

        b+=1
    
        return mincut
    
#error check - why variable passing to this function isn't working ...
```
